```python
import matplotlib.pyplot as plt
import math
import numpy as np
import scipy as sp
from matplotlib import cm, colors
from mpl_toolkits.mplot3d import Axes3D
import random
#n=number of points,d=dimension

def samplepoints(n,d):
    point= np.random.normal(0, 1,n*d)
    point=np.reshape(point, (n,d))
    for x in point:
        r=np.random.uniform(0, 1)
        R=d*np.power(r,d-1)
        x/=np.linalg.norm(x)    
        x*=R
    return point;


#plot points in 2d
fig = plt.figure()

point = samplepoints(1000, 2);
sample2d_x=[]
sample2d_y=[]
for pt in point:
    sample2d_x.append(pt[0])
    sample2d_y.append(pt[1])

axes = plt.gca()
axes.set_xlim([-2,2])
axes.set_ylim([-2,2])
plt.scatter(sample2d_x,sample2d_y,s=4);


#plot points in 3d
sample3d_X=[]
sample3d_Y=[]
sample3d_Z=[]
point = samplepoints(1000, 3);
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for pt in point:
    sample3d_X.append(pt[0])
    sample3d_Y.append(pt[1])
    sample3d_Z.append(pt[2])
ax.scatter(sample3d_X, sample3d_Y,sample3d_Z,s=4)


    
```
