Plot the three functions and the relative volume of the $\frac{1}{d}$-annulus for $d \in \{1, ...,20\}$

In previous exercises we found that $\frac{vol(A_{\varepsilon}^{d})}{vol(B^{d})} = 1 - (1 - \varepsilon)^{d} \geq 1 - e^{-\varepsilon \cdot d}$. We define a function *ratio* that calculates the ratio for a given $\varepsilon$ and *d* and also handles the exact formula and the approximation.

```python
def ratio(eps, dim, approx=False):
    if not approx:
        return 1-(1-eps)**dim
    else:
        return 1-math.exp(-eps*dim)
```

We then proceed to plot the *ratio* for $\varepsilon \in \{\frac{1}{100}, \frac{1}{\sqrt{d}}, \frac{1}{d^{2}}, \frac{1}{d}\}$ and $d \in \{1, ...,20\}$

```python
# dimension range.
r = range(1,21)

# eps = 1 / 100.
plt.plot([ratio(1/100, dim) for dim in r], label='1/100')
plt.plot([ratio(1/100, dim, approx=True) for dim in r], label='1/100, approx')

# eps = 1 / sqrt(dimension).
plt.plot([ratio(1/math.sqrt(dim), dim) for dim in r], label='1/sqrt(dim)')
plt.plot([ratio(1/math.sqrt(dim), dim, approx=True) for dim in r], label='1/sqrt(dim), approx')

# eps = 1 / (dimension^2).
plt.plot([ratio(1/(dim**2), dim) for dim in r], label='1/(dim**2)')
plt.plot([ratio(1/(dim**2), dim, approx=True) for dim in r], label='1/(dim**2) , approx')

# eps = 1 / dimension.
plt.plot([ratio(1/dim, dim) for dim in r], label='1/dim')
plt.plot([ratio(1/dim, dim, approx=True) for dim in r], label='1/dim, approx')

# aesthetics. Using matplotlib-label-lines.
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.xlabel("dimension d", fontsize=16)
plt.ylabel("volume of annulus / volume of unit ball", fontsize=16)
labelLines(plt.gca().get_lines(),zorder=2.5, fontsize=12)
fig.savefig("E3A2iv.ps", format="ps")

plt.show()
```
